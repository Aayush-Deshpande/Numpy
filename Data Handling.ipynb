### Handling data in matrices

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Saving and loading Arrays

```python
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])
array3 = np.array([11, 12, 13, 14, 15])

np.save('array1.npy', array1)
np.save('array2.npy', array2)
np.save('array3.npy', array3)

loaded_array1 = np.load('array1.npy')
loaded_array2 = np.load('array2.npy')
loaded_array3 = np.load('array3.npy')

print("Array1: {}\nArray2: {}\nArray3: {}".format(loaded_array1, loaded_array2, loaded_array3))
```

## Loading images and dark mode

```python
try:
    logo = np.load('numpy-logo.npy')
    plt.figure(figsize=(10, 5))
    plt.title("Numpy Logo")
    plt.grid(False)
    plt.subplot(121)
    plt.imshow(logo)
except FileNotFoundError:
    print("=== FileNotFoundError ===")
```

## Convert image to Dark Mode

```python
dark_logo = 1 - logo
try:
    plt.figure(figsize=(10, 5))
    plt.title("Numpy Dark Logo")
    plt.grid(False)
    plt.subplot(121)
    plt.imshow(dark_logo)
except FileNotFoundError:
    print("=== FileNotFoundError ===")
```
