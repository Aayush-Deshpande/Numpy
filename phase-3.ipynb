# Advance operation with Business examples


```python
import numpy as np
import matplotlib.pyplot as plt
```

### Data_Structure: [restaurant_id, 2021, 2022, 2023, 2024]
#### Print Specefic Part


```python
sales_data = np.array([[1, 150000, 180000, 220000, 100000], # Paradise Biryani 
                      [2, 120000, 140000, 160000, 190000], # Beijing Bytes
                      [3, 200000, 230000, 260000, 300000], # Pizza Hut
                      [4, 180000, 210000, 50000, 270000], # Burger Point
                      [5, 160000, 185000, 205000, 230000], # Chai Point
                      ])
print("=== Zomato Sales Analysis ===\n Shape is: ", sales_data.shape)
print("Sample data for first 3 Restaurants:\n", sales_data[:3])
```

```python
print("Column wise sales: ", np.sum(sales_data[:,1:], axis=0))
print("Row wise sales: ", np.sum(sales_data[:,1:], axis=1))
```

#### Min sales per Restaurant

```python
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Min. sales per rest: ", min_sales)
max_sales = np.max(sales_data[:, 1:], axis=1)
print("Max. sales per rest: ", max_sales)
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Avg. sales per rest: ",avg_sales)
tot_sales = np.sum(sales_data[:, 1:], axis=0)
print("Total sales per year: ", tot_sales)
cum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative Sales per yr:\n", cum_sales)
```

## Plot Graph

```python
plt.figure(figsize=(5, 5))
plt.plot(np.mean(cum_sales, axis=0))
plt.title("Average cumulative sale across all restaurants:")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show
```

### Vector Operations 

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])
print("Vector addition: ", (vector1 + vector2))
print("Vector multiplication: ", (vector1 * vector2))
print("Dot Product: ", np.dot(vector1, vector2))
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between vectors: X = {} radians" .format(angle))

```

### Character Vectors

```python
restaurant_types = np.array(['Biryani', 'Maggi', 'Chicken', 'Eggs'])
vectorized_done = np.vectorize(str.upper)
print("Vectorized: ", vectorized_done(restaurant_types))
print("Monthly sales:\n", (sales_data[:, 1:] / 12))
```
